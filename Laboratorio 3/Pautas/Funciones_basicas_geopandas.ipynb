## Importación

````python
import geopandas as gpd
````

## Abrir un archivo Shape (SHP)

`gpd.read_file()`

````python
folder_path = 'shapes/distritos censales'
file_path = 'Distritos Censales con GSE.shp'
path = folder_path + '/' + file_path
path
````

````python
distritos = gpd.read_file(path)
````

````python
distritos
````

## Visualizar los datos del DBF

````python
distritos.crs
````

## Acceder a la información del DBF
`gdf.loc[]`

````python
distritos.head(1)
````

````python
distritos['CODIGO']
````

````python
distritos[['CODIGO', 'E']]
````

````python
distritos['E'][4]
````

````python
# loc -> filas, columnas
distritos.loc[4:10,'E':'ABC1']
````

````python
distritosE = distritos.loc[distritos['E']>100]
distritosE.head()
````

````python
distritosE['E'][5]
````

## Visualizar las figuras del SHP
`gdf.plot()`

````python
distritos.plot(figsize = (40,40), edgecolor = 'red')
````

````python
distritosE.plot(figsize = (40,40), edgecolor = 'red')
````

````python
base = distritos.plot(figsize = (40,40), edgecolor = 'red', color = 'skyblue')
distritosE.plot(ax = base, figsize = (40,40), edgecolor = 'red')
````

## Guardar nuestro trabajo
`to_file()`

````python
distritosE.to_file('distritos con mas E.shp')
````
